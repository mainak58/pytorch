```python
import torch
import torchvision
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# -------- Step 1: Load Model --------
# Example: Using FasterRCNN with ResNet backbone (common for object detection)
# You need to adjust this part based on the original training code

num_classes = 2  # 1 class (chart) + background
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=False, num_classes=num_classes)

# Load your weights
checkpoint = torch.load(r"C:\Users\MAINAK\Desktop\opencv\chart_detector.pth", map_location="cpu")
model.load_state_dict(checkpoint)
model.eval()

img_path = "nc_page_67.png"  
image = Image.open(img_path).convert("RGB")

# Preprocess (convert to tensor, normalize)
transform = transforms.Compose([
    transforms.ToTensor()
])
img_tensor = transform(image).unsqueeze(0)  # add batch dimension

# -------- Step 3: Run Inference --------
with torch.no_grad():
    predictions = model(img_tensor)

# Predictions contain boxes, labels, scores
boxes = predictions[0]['boxes']
scores = predictions[0]['scores']

# -------- Step 4: Draw Results --------
img_cv = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)

for i, box in enumerate(boxes):
    if scores[i] > 0.5:  # confidence threshold
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(img_cv, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(img_cv, f"Chart {scores[i]:.2f}", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

# -------- Step 5: Show Result --------
plt.imshow(img_cv)
plt.axis("off")
plt.show()

# Or save
cv2.imwrite("result.png", cv2.cvtColor(img_cv, cv2.COLOR_RGB2BGR))
print("Result saved as result.png")
```
